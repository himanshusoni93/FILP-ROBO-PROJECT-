## Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps


```python
import folium ## to map coordinates data on a map
    
## Import the all basic use full liblrary 
import re
import selenium
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException,ElementNotInteractableException
## lets connect with web driver 

driver = webdriver.Chrome(r"C:/Users/Acer/chromedriver/chromedriver.exe")

## spacifiying the url

Url="https://www.google.co.in/maps"
## lets open the web page through our web page 
driver.get(Url)

## check the All input box by using ID 
city=str(input("Enter the city name "))
inputboxs = driver.find_element_by_id("searchboxinput")
inputboxs.clear()
#print(len(inputboxs))
time.sleep(3)
inputboxs.send_keys(city)                                                     # entering values in search bar
button = driver.find_element_by_id("searchbox-searchbutton")               # locating search button
button.click()                                                             # clicking search button
time.sleep(4)
try:
    url_string = driver.current_url
    print("URL Extracted: ", url_string)
    lat_lng = re.findall(r'@(.*)data',url_string)
    if len(lat_lng):
        lat_lng_list = lat_lng[0].split(",")
        if len(lat_lng_list)>=2:
            lat = lat_lng_list[0]
            lng = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(lat, lng))

except Exception as e:
        print("Error: ", str(e))
```

## Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 – September 20) from trak.in.

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd

## spacifiying the url
URL = "https://trak.in/india-startup-funding-investment-2015/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
tables = soup.find_all("table")
```

```python
## scrap the tabuler data of the July Month
July_Month_table=tables[6]
July_tab_data = [[cell.text for cell in row.find_all(["th","td"])]
                               for row in July_Month_table.find_all("tr")]
July_df = pd.DataFrame(July_tab_data)
##Sep_df.head()
#To move the first row to the headers, simply type
July_df.columns = July_df.iloc[0,:]
July_df.drop(index=0,inplace=True)
July_df.head()
```

```python
## scrap the tabuler data of the August Month
August_Month_table=tables[5]
August_tab_data = [[cell.text for cell in row.find_all(["th","td"])]
                        for row in August_Month_table.find_all("tr")]
Aug_df = pd.DataFrame(August_tab_data)
##Sep_df.head()
#To move the first row to the headers, simply type
Aug_df.columns = Aug_df.iloc[0,:]
Aug_df.drop(index=0,inplace=True)
Aug_df.head()
```

```python
September_month_table=tables[4]
Sep_tab_data = [[cell.text for cell in row.find_all(["th","td"])]
                        for row in September_month_table.find_all("tr")]
Sep_df = pd.DataFrame(Sep_tab_data)
##Sep_df.head()
#To move the first row to the headers, simply type
Sep_df.columns = Sep_df.iloc[0,:]
Sep_df.drop(index=0,inplace=True)
Sep_df.head()
```

```python
funding_deals_df = pd.concat([July_df,Aug_df,Sep_df], keys=['JULY','AUGUST','SEPTEMBER'])
funding_deals_df
```
